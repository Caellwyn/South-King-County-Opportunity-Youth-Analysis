```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import pandas as pd
import math
import branca as cm

%matplotlib inline
```

```python
pop_dic = {'11604': 867.0,
 '11610': 699.0,
 '11611': 982.0,
 '11612': 1360.0,
 '11613': 1052.0,
 '11614': 964.0,
 '11615': 737.0}
```

```python
reg = {'11612': ['Federal Way,Des moines Cities & Vashon island',  pop_dic['11612'], 47.3223, -122.3126, 110 ],
 '11604': ['South East Seattle',  pop_dic['11604'], 47.5923, -122.3022, 30.5],
 '11615': ['Maple Valley, Covingtion and Enumclaw Cities',  pop_dic['11615'], 47.3903, -122.0454, 18],
 '11614': ['Auburn City & Lakeland',  pop_dic['11614'], 47.3073, -122.2285, 35],
 '11613': ['Kent City',  pop_dic['11613'], 47.3809, -122.2348, 34.43],
 '11611': ['Burien, SeaTac, Tukwita Cities',  pop_dic['11611'], 47.4668, -122.3405, 31],
 '11610': ['Renton City, Fairwood, Bryn Mawr and Skyway', pop_dic['11610'], 47.4797, -122.2079, 31]}
```

```python
name = [i[0] for i in reg.values()]
lat = [i[2] for i in reg.values()]
lon = [i[3] for i in reg.values()]
population = [i[1] for i in reg.values()]
area = [i[4] for i in reg.values()]
```

```python
m = folium.Map(location=[47.45,-122.3321], zoom_start=11)


for i in range(0,len(reg.values())):
   folium.CircleMarker(
      location=[lat[i], lon[i]],
      popup= name[i] + ', \nPopulation:' + str(population[i]),
      radius=population[i]/14,
      color='blue',
      fill=True,
      fill_color='blue'
   ).add_to(m)

colormap = cm.colormap.LinearColormap(colors=['blue','blue','blue','blue','blue'], index=[0,600,1200,1800,2400],vmin=0,vmax=2400);
colormap.caption = 'Population of Oppurtunity Youth';
colormap.add_to(m);
```
